```python
import pandas as pd
import os
import shutil

devf = '/content/gdrive/MyDrive/dev.csv'
devcsv = pd.read_csv(devf)
dfdev = pd.DataFrame(devcsv)
#dftrain = dfdev['split']=='train'
dftrain = dfdev[dfdev['split'] == 'train'].copy()
dftest = dfdev[dfdev['split'] == 'val'].copy()
#dftrain2 = dftrain[dftrain.labels.str.contains('home',case=False)]
#dftrain_home = pd.DataFrame(dftrain2)
dftrain.info()
dftrain.head()

dftest.info()
dftest.head()
```

```python
dev_png_dir = '/content/gdrive/MyDrive/dumptrainpng2_new/'
target_dir = '/content/gdrive/MyDrive/dataset/train'
dftrain["figfiles"] = dftrain.apply(lambda row: dev_png_dir+str(row.fname)+'.png', axis=1)
dftest["figfiles"] = dftest.apply(lambda row: dev_png_dir+str(row.fname)+'.png', axis=1)
# function to check if directory exists

filect = 0
for i in dftrain.itertuples():
    # get labels for audio
    labels = i[2]
    #label_part = labels.split(':')[0]
    labels_part1 = labels.partition(',')[0]
    labels_part2= labels_part1.replace("(", "")
    labels_part = labels_part2.replace(")", "")
    #print(labels_part)
    targetdir_chk = target_dir+'/'+labels_part
    if os.path.isdir(targetdir_chk) == False:
        os.mkdir(targetdir_chk)
        print(targetdir_chk+' directory does not exist so it was created')
    #print(i[1])
    src_path=i[5]
    dst_path=targetdir_chk+'/'+str(i[1])+'.png'
    if os.path.isfile(src_path) == True:
        print('Moving '+ src_path +' to '+dst_path)
        shutil.move(src_path, dst_path)
    else: print('Error: File '+src_path+' is missing. Unable to move!')



```
