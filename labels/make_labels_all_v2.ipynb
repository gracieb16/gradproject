```python
import pandas as pd
import os
import shutil

traintestf = '/content/gdrive/MyDrive/dev.csv'
validf = '/content/gdrive/MyDrive/eval.csv'
devcsv = pd.read_csv(traintestf)
dfdev = pd.DataFrame(devcsv)

#validation dataset
dfval2 = pd.read_csv(validf)
dfval = pd.DataFrame(dfval2)

# Training Dataset
dftrain2 = dfdev[dfdev['split'] == 'train'].copy()
dftrain = pd.DataFrame(dftrain2)

# Test Dataset
dftest2 = dfdev[dfdev['split'] == 'val'].copy()
dftest = pd.DataFrame(dftest2)

dftrain.info()
dftrain.head()

dftest.info()
dftest.head()

dfval.info()
dfval.head()
```

```python
dev_png_dir = '/content/gdrive/MyDrive/dumptrainpng2_new/'
val_png_dir = '/content/gdrive/MyDrive/dumpvalidpng/'
target_dir = '/content/gdrive/MyDrive/dataset/train'

dftrain["figfiles"] = dftrain.apply(lambda row: dev_png_dir+str(row.fname)+'.png', axis=1)
dftest["figfiles"] = dftest.apply(lambda row: dev_png_dir+str(row.fname)+'.png', axis=1)
dfval["figfiles"] = dfval.apply(lambda row: val_png_dir+str(row.fname)+'.png', axis=1)

```

```python
# work on training spectograms
target_dir = '/content/gdrive/MyDrive/dataset3/train'
filect = 0
for i in dftrain.itertuples():
    # get labels for audio
    labels = i[2]
    labels_part1 = labels.partition(',')[0]
    labels_part2= labels_part1.replace("(", "")
    labels_part = labels_part2.replace(")", "")
    #print(labels_part)
    targetdir_chk = target_dir+'/'+labels_part
    if os.path.isdir(targetdir_chk) == False:
        #os.mkdir(targetdir_chk)
        print(targetdir_chk+' directory does not exist so it was created')
    #print(i[1])
    src_path=i[5]
    dst_path=targetdir_chk+'/'+str(i[1])+'.png'
    print(dst_path)
    if os.path.isfile(src_path) == True:
      filect=filect+1
      print('Moving file count '+str(filect)+ src_path +' to '+dst_path)
      #shutil.move(src_path, dst_path)
    else: print('Error!: File '+src_path+' is missing. Unable to move!')


```
