```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
import pandas as pd
import time
devf = '/content/gdrive/My Drive/dev.csv'
#devf = '/Users/gracie/ML/AEA_dataset/freesound/FSD50K.ground_truth/dev.csv'
devcsv = pd.read_csv(devf)
dfdev = pd.DataFrame(devcsv)
#dftrain = dfdev['split']=='train'
dftrain = dfdev[dfdev['split'] == 'train'].copy()
print(dftrain.info())
print(dftrain.head())
```

```python

dev_audio_dir = '/content/gdrive/MyDrive/devwav/'
dev_png_dir = '/content/gdrive/MyDrive/trainpng/'
dftrain["wavfiles"] = dftrain.apply(lambda row: dev_audio_dir+str(row.fname)+'.wav', axis=1)
dftrain["figfiles"] = dftrain.apply(lambda row: dev_png_dir+str(row.fname)+'.png', axis=1)

dftrain.info()
dftrain.head()
startime = time.ctime()
print(time.ctime())
srhz=44100
frame_size = int(srhz/1000*960)
frame_shift = int(srhz/1000*25)
```

```python
for i in dftrain.itertuples():
    file = i[5]
    pngfile = i[6]
    print('Working on '+file)
    signal, sr = librosa.load(file, sr=srhz)
    stft = librosa.stft(signal, n_fft=frame_size, hop_length=frame_shift)
    # This will yield 96-length spectrograms
    spectrogram = np.abs(stft)
    # converted to 64 bin log-scaled Mel spectrogram,
    melspec = librosa.feature.melspectrogram(y=signal, sr=sr, S=spectrogram, n_fft=frame_size, hop_length=frame_shift, win_length=None, window='hann', center=True, pad_mode='reflect', n_mels=64)
    melspec_db = librosa.power_to_db(melspec, ref=np.max)
    fig = librosa.display.specshow(melspec_db, hop_length=frame_shift)
    fig2 = plt.gcf()
    fig2.set_size_inches(2.24, 2.24)
    plt.savefig(pngfile,dpi=100)

#enddate = datetime.datetime.today()
#print enddate.strftime("%Y%m%d %H:%M")
print('start time: ')
print(startime)
print(time.ctime())
```
